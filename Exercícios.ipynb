# Parte 3 - Aula 3

```python
#a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2) - np.sum(X)**2)
#b = np.mean(Y) - a*np.mean(X)
```

```python
# Coeficiente angular e linear para laranja:
```

```python
diametro_laranja #eixo x
peso_laranja #eixo y

n = peso_laranja.size
X = np.copy(diametro_laranja)
Y = np.copy(peso_laranja)

a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2) - np.sum(X)**2)
b = np.mean(Y) - a*np.mean(X)
```

```python
y = a * X + b
```

```python
np.linalg.norm(Y - y)
```

```python
plt.plot(diametro_laranja, peso_laranja)
plt.plot(diametro_laranja, y)
plt.show()
```

```python
# Coeficiente angular e linear para toranja:
```

```python
diametro_toranja #eixo x
peso_toranja #eixo y

n = peso_toranja.size
X = np.copy(diametro_toranja)
Y = np.copy(peso_toranja)

a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2) - np.sum(X)**2)
b = np.mean(Y) - a*np.mean(X)
```

```python
y = a * X + b
```

```python
np.linalg.norm(Y - y)
```

```python
plt.plot(diametro_toranja, peso_toranja)
plt.plot(diametro_toranja, y)
plt.show()
```

```python
b = 17 #Coeficiente linear
```

```python
np.random.seed(84)
coef_angular = np.random.uniform(low = 0.0, high = 30.0, size = 100)
```

```python
coef_angular
```

```python
norma = np.array([])

for i in range(100):
    norma = np.append(norma, np.linalg.norm(peso_laranja - (coef_angular[i] * diametro_laranja + b)))
```

```python
dados_laranja = np.column_stack([norma, coef_angular])
```

```python
np.savetxt('dados_laranja.csv', dados_laranja, delimiter = ',')
```

##### Sua missão é construir uma função que recebe dois parâmetros, a coluna que você quer analisar se tem outlier e o DataFrame em questão. A intenção de fazer isso é aproveitar novamente essa função tanto para outras colunas quanto para outros DataFrames futuramente em outras análises. O método que deve ser implementado para identificação do outlier é do intervalo interquartil:

```python
#limite inferior = q1 - 1.5 * iqr
# limmite superior = q3 + 1.5 * iqr
# iqr = q3 - q1
# df[coluna].quantile(.25 ou .75)

def verifica_outlier(dataframe, coluna):
    Q1 = dataframe[coluna].quantile(.25)
    Q3 = dataframe[coluna].quantile(.75)

    IQR = Q3 - Q1

    limite_inferior = Q1 - (1.5 * IQR)
    limite_superior = Q3 + (1.5 * IQR)

    filtro_outlier  = (dataframe[coluna] < limite_inferior) | (dataframe[coluna] > limite_superior)

    dataframe_copia = dataframe.copy()

    outliers = dataframe_copia[filtro_outlier]

    df_sem_fraude = dataframe_copia[~filtro_outlier]
    df_sem_fraude.reset_index(drop = True, inplace = True)
    
    return outliers, df_sem_fraude

fraudes, df_valido = verifica_outlier(df, 'Valor da transação')

```

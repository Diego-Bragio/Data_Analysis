# Aula 1

```python
import pandas as pd
```

```python
df = pd.read_csv('imigrantes_canada.csv')
```

```python
df
```

```python
df.info()
```

```python
#Alterando o índice do Dataframe:
df.set_index('País', inplace = True)
```

```python
df
```

```python
#Sequência de anos:
anos = list(map(str, range(1980, 2014)))
anos
```

```python
#Coletando dados do Brasil no DataFrame df:
brasil = df.loc['Brasil', anos]
brasil
```

```python
#Criando DataFrame 'dados_brasil' a partir do dicionário 'brasil_dict':
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil
```

```python
#Importando matplotlib:
import matplotlib.pyplot as plt
```

```python
#Plotando gráfico:
plt.figure(figsize = (8, 4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.xticks(list(map(str, range(1980, 2011, 5))))
plt.yticks(list(map(int, range(500, 2501, 500))))
plt.title('Imigração Brasil - Canadá entre 1980 e 2013')
plt.xlabel('Ano')
plt.ylabel('Número de Imigrantes')
plt.show()
```

```python
#Criando figura:
fig, ax = plt.subplots(figsize = (8, 4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
ax.set_title('Imigração do Brasil para o Canadá\n1980 - 2013')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_xlabel('Ano')
ax.set_ylabel('Número de Imigrantes')

plt.show()
```

```python
#Gerando mais de um gráfico em uma mesma figura:
fig, axs = plt.subplots(nrows = 1, ncols = 2, figsize = (15, 5))

axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Imigrantes')
axs[0].set_title('Imigração do Brasil para o Canadá\n1980 - 2013')
axs[0].grid()

axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Imigrantes')
axs[1].set_title('Boxplot de Imigração do Brasil para o Canadá\n1980 - 2013')
axs[1].grid()

plt.show()
```

```python
#Checando estatísticas descritivas para o DataFrame:
dados_brasil.describe()
```

```python
#Criando plots multidirecionais:
fig, axs = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 6)) #Tamanho é para CADA plot
fig.subplots_adjust(hspace = 0.5 , wspace = 0.3)
fig.suptitle('Imigração dos quatro maiores paídes da América do Sul para o Canadá de 1980 à 2013')

#Brasil:
axs[0, 0].plot(df.loc['Brasil', anos])
axs[0, 0].set_title('Brasil\nImigração para o Canadá (1980 - 2013)')

#Colômbia:
axs[0, 1].plot(df.loc['Colômbia', anos])
axs[0, 1].set_title('Colômbia\nImigração para o Canadá (1980 - 2013)')

#Argentina:
axs[1, 0].plot(df.loc['Argentina', anos])
axs[1, 0].set_title('Argentina\nImigração para o Canadá (1980 - 2013)')

#Peru:
axs[1, 1].plot(df.loc['Brasil', anos])
axs[1, 1].set_title('Peru\nImigração para o Canadá (1980 - 2013)')

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))

for ax in axs.flat:
    ax.set_xlabel('Ano')
    ax.set_ylabel('Imigrantes')

ymin = 0
ymax = 7000

for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)



plt.show()
```

```python
#Trabalhando com cores:
df.head(1)
```

```python
america_sul = df.query('Região == "América do Sul"')
america_sul
```

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal',  'tomato']

fig, ax = plt.subplots(figsize = (12, 5))
ax.bar(america_sul.index, america_sul['Total'], color = cores)
ax.set_title('Imigração América do Sul - Canadá:\n1980 - 2013\nBrasil foi o 4° país com maior número de imigrantes', loc = 'center', fontsize = 18)
ax.set_xlabel('') #Em branco
ax.set_ylabel('Imigrantes', fontsize = 14)
ax.xaxis.set_tick_params(labelsize = 8)
ax.yaxis.set_tick_params(labelsize = 8)
plt.show()
```

```python
#Ordenando DataFrame para gerar visualização crescente/decrescente
america_sul_sorted = america_sul.sort_values('Total', ascending = True)
america_sul_sorted.head()
```

```python
#Alterando orientação e alguns parâmetros de cor do gráfico:
#cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal',  'tomato']
cores = []

for pais in america_sul_sorted.index:
    if pais == 'Brasil':
        cores.append('green')
    else:
        cores.append('silver')

fig, ax = plt.subplots(figsize = (12, 5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color = cores)
ax.set_title('Imigração América do Sul - Canadá:\n1980 - 2013', fontsize = 18)
ax.set_xlabel('Imigrantes', fontsize = 14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize = 8)
ax.yaxis.set_tick_params(labelsize = 8)

for i, v in enumerate(america_sul_sorted['Total']): #iterando para cada valor numérico da Series
    ax.text(v + 20, i, str(v), color = 'black', fontsize = 10, ha = 'left', va = 'center') #Posicionamento textual

ax.set_frame_on(False) #Remove frame

ax.get_xaxis().set_visible(False)

ax.tick_params(axis = 'both', which = 'both', length = 0)

plt.show()
```

```python
#Seaborn:
import seaborn as sns
```

```python
sns.set_theme() #gera o tema padrão do seaborn
```

```python
top_10 = df.sort_values('Total', ascending = False).head(10) #Criando um DataFrame com os 10 maiores índices de imigração
top_10
```

```python
sns.barplot(data = top_10, x = top_10.index, y = 'Total', palette = 'tab10') #Plotando gráfico de barras verticais
#Para gerar um gráfico igual ao modelo proposto, foi necessário informar o parâmetro 'palette'
plt.show()
```

```python
sns.barplot(data = top_10, x = 'Total', y = top_10.index, orient = 'h', palette = 'tab10')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8, 4))
ax = sns.barplot(data = top_10, x = 'Total', y = top_10.index, orient = 'h', palette = 'tab10')
#ax.set(title = 'Maiores imigrações: Canadá\n1980 - 2013', xlabel = 'Imigrantes', ylabel = '') #ajustar diversos parâmetros de uma única vez
ax.set_title('Maiores Imigrações: Canadá\n1980 - 2013', loc = 'left', fontsize = 16)
ax.set_xlabel('Imigrantes', fontsize = 14)
ax.set_ylabel('')
plt.show()
```

```python
#Criando função para testar paletas:
def gerar_grafico_paleta(palette):
    fig, ax = plt.subplots(figsize = (8, 4))
    ax = sns.barplot(data = top_10, x = 'Total', y = top_10.index, orient = 'h', palette = palette)
    ax.set_title('Maiores imigrações: Canadá\n1980 - 2013', fontsize = 16, loc = 'left')
    ax.set_xlabel('Imigrantes', fontsize = 14)
    ax.set_ylabel('')
    plt.show()
    return ax
```

```python
gerar_grafico_paleta('Blues')
plt.show()
```

```python
gerar_grafico_paleta('Blues_r') #_r inverte as cores do gráfico
plt.show()
```

```python
gerar_grafico_paleta('rocket')
plt.show()
```

```python
gerar_grafico_paleta('coolwarm')
plt.show()
```

```python
gerar_grafico_paleta('tab10')
plt.show()
```

```python
#Alterando temas dos gráficos:
sns.set_theme(style = 'dark')
gerar_grafico_paleta('tab10')
plt.show()
```

```python
sns.set_theme(style = 'whitegrid')
gerar_grafico_paleta('tab10')
plt.show()
```

```python
sns.set_theme(style = 'white')
gerar_grafico_paleta('tab10')
plt.show()
```

```python
sns.set_theme(style = 'ticks')
gerar_grafico_paleta('tab10')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8, 4))
ax = sns.barplot(data = top_10, x = 'Total', y = top_10.index, orient = 'h', palette = 'tab10')
ax.set_title('Maiores Imigrações: Canadá\n1980 - 2013', fontsize = 18, loc = 'left')
ax.set_xlabel('Imigrantes', fontsize = 14)
ax.set_ylabel('')
sns.despine() #Remove linhas a direita e superior da caixa
plt.show()
```
